## Song dictionary
We will create a simple dictionary to store song names, albums and lyrics.

```python
song_dict = {}
for song in data['songs']:
    
    song_dict['name'] = song['full_title']

    song_dict['lyrics'] = song['lyrics']

    song_dict['album'] = song['album']['name']
```

## Generate files for each song

```python
song_dict =  {}
for song in artist.songs:

    try: 
        #Retrieve song name
        pattern = r'"([^"]*)"'
        song_name = re.findall(pattern, song.lyrics)[0].lower()

        print('The "{}" song is being retrieved.'.format(song_name))

        #Extracting the lyrics for the song
        pattern = '\n(.*)'
        lines_list = re.findall(pattern, song.lyrics) #retrieving only the lyrics and song sections
        text = ' '.join(lines_list).lower() #joining all lines into a single string
        text = re.sub('\[.*?\]', "", text) #remove song section names
        text = re.sub("-ohyou might also likeembed", "", text) #removing the 'ohyou might also likeembed' comment at the end
        text = re.sub("  ", " ", text) #removing double spaces
        
        #Save lyrics in a file named after the song
        text_file = open(song_name+".txt", "w")
        n = text_file.write(text)
        text_file.close()

        #Creating the dictionary of songs
        song_dict[song_name] = text

        print('The "{}" song was stored in the dictionary.'.format(song_name))
    
    except:
        print(song)
        
```
